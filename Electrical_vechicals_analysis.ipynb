```python
# df["Market Share"]=df["Market Share"].str.replace("%","")
# df["Market Share"]=df["Market Share"].astype(float)
# df.drop("S no",axis=1,inplace=True)
# print(df.isnull().sum())
# Remove % sign
df["Market Share"] = df["Market Share"].astype(str)
df["Market Share"] = df["Market Share"].str.replace("%","")

df["Market Share"] = df["Market Share"].astype(float)

# Remove Serial Number

df.drop("S No", axis=1, inplace=True)

# Check missing values

print(df.isnull().sum())
```

```python
total_units = df["Total"].sum()
print(f"\nTotal Industry sales:{total_units:,} units")

# total_units = df["Total"].sum()
# print(f"\nTotal Industry Sales: {total_units:,} units")
```

```python
Months = ["Apr","May","Jun","Jul","Aug","Sep",
          "Oct","Nov","Dec","Jan","Feb","March"]
df["Average Sales"]=df[Months].mean(axis=1)
df["Maximun Sales"]=df[Months].max(axis=1)
df["Minimun Sales"]=df[Months].min(axis=1)
df["Sales td"] = df[Months].std(axis=1)
```

```python
# import matplotlib.pyplot as plt
# plt.figure(figsize(10,10))
# plt.bar(df=["Maker"], df["Total"])
# plt.xticks(rotation=90)
# plt.title("Total EV Sales")
# plt.show()

import matplotlib.pyplot as plt

# Fixed: Added '=' inside figsize and corrected how DataFrame columns are passed
plt.figure(figsize=(10, 10))
plt.bar(df["Maker"], df["Total"])
plt.xticks(rotation=90)
plt.title("Total EV Sales")
plt.show()
```

```python
df["Growth"] = ((df["March"] - df["Apr"]) / df["Apr"]) * 100
```

```python
import numpy as np

df["Growth"] = np.where(
    df["Apr"] == 0,
    np.nan,
    ((df["March"] - df["Apr"]) / df["Apr"]) * 100
)
```

```python
plt.figure(figsize=(12,6))

plt.bar(
df["Maker"],
df["Growth"]
)

plt.xticks(rotation=90)

plt.title("Sales Growth (%)")

plt.show()
```
